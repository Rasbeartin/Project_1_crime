# LA Crime Data

* Read the crime data for LA from https://data.lacity.org/A-Safe-City/Crime-Data-from-2010-to-Present/y8tr-7khq/data
* Take the data from year 2015-present
* Formatted the data with Type of crime, Date Occurred and crime count
* Store the formatted data in csv file

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python

# File to Load
data_folder = 'C:/data/crime_date_raw/Los_Angeles_city/'
raw_file = data_folder + 'crime_data_LA_2015.csv'

# Read the File and store into Pandas Data Frames
la_df = pd.read_csv(raw_file)
la_df.head()
```

```python
# Read the values from Crime Code Description
cats = la_df['Crime Code Description'].unique()

cats_df = pd.DataFrame({'Crime Code Description' : cats})
#Write the Crime Code Description in csv and then work in excel to group them by primary type
cats_df.to_csv('categories.csv')
```

Edit categories_key.csv

```python
#Read the edited csv for the final category type
cat_keys = pd.read_csv('categories_key.csv')
# Mearge the original dataframe with the new categories key 
la_short_cats = pd.merge(la_df, cat_keys, on='Crime Code Description')
```

```python
la_short_cats.head()
```

```python
crimes = la_short_cats['Categories Final'].unique()

# Add columns for each crime category, 0 = crime is not this category, 1 = crime is this category
for c in crimes:
    la_short_cats[c] = la_short_cats['Categories Final'].apply(lambda s: int(s==c))
```

```python
# Group by date and export the counts to csv
LA_by_date = la_short_cats.groupby(['Date Occurred'])
LA_by_date[crimes].sum().to_csv('LA_by_date.csv')
```
