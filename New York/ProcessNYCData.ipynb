## New York City  Crime Data
Read the crime data for Seattle from https://data.cityofnewyork.us/Public-Safety/NYPD-Complaint-Data-Current-Year-To-Date-/5uac-w243/data
Take the data from year 2015-present
Formatted the data with Type of crime, Date Occurred and crime count
Store the formatted data in csv file

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load
csv_path = ("../../Data_folder/NYPD_Crime_data_2015.csv")
data_file = pd.read_csv(csv_path)
data_file.head()
```

```python
len(data_file)
```

```python
data_file["CMPLNT_FR_DT"]= data_file["CMPLNT_FR_DT"].astype('datetime64[ns]')
```

```python
#Read all the categories of crime 
cats=data_file["OFNS_DESC"].unique()
cats_df = pd.DataFrame({"OFNS_DESC":cats})
cats_df.to_csv('categories.csv')

#Go to the excel and edit the file
```

```python
#read the new csv file with final category colummn and add to our new dataframe
cats_key = pd.read_csv("category_keys.csv")
cats_key.head()
```

```python
# Merge new column to the original data set
nyc_data = pd.merge(data_file,cats_key, on = "OFNS_DESC", how = "left")
```

```python
#Read all the final categories of crimes
crimes = nyc_data["Category Final"].unique()
```

```python
# Add columns for each crime category, 0 = crime is not this category, 1 = crime is this category
for c in crimes:
    nyc_data[c] = nyc_data["Category Final"].apply(lambda s: int(s==c))
# Group by date and export the counts to csv
nyc_final_df = nyc_data.groupby(["CMPLNT_FR_DT"])
nyc_final_df[crimes].sum().to_csv("NYC_by_date.csv")
```
