### Define a function to convert date to mm/dd/yyyy format using datetime import. If data set provides time of crime, format time to HM

```python
def format_date(s):
    return datetime.datetime.strptime(s, "%m/%d/%Y %H:%M")

sample_city_renamed.dropna(inplace=True)
sample_city_renamed['timestamp'] = sample_city_renamed['Crime Date'].apply(format_date)

sample_city_renamed['Hour'] = sample_city_renamed['timestamp'].apply(lambda t: t.hour)
sample_city_renamed['Date'] = sample_city_renamed['Crime Date'].apply(lambda t: t.split(' ')[0])
```

### Preview data with new date format

```python
sample_city_renamed.head()
```

### Use a lambda function to get the count of crime committed per crime type per day
* Group dataframe by date using 'groupby'
* Preview dataframe with crime committed per type by date
* Note for very large data sets, data will be split up by year then merged back to one file after functions have been performed

```python
crimes = sample_city_renamed['Cateogry Crime Type'].unique()

Add columns for each crime category, 0 - crime is not this category, 1 = crime is this cateogry

for c in crimes:
    sample_city_renamed = sample_city_renamed['Cateogry Crime Type'].apply(lambda s: int(s==c))

sample_city_renamed = sample_city_renamed.groupby(['Date'])

sample_city_renamed[crimes].sum().head()
```

### Output final city csv to be used for analysis and graph plotting

```python
output_file = "sample_city_renamed.csv"
sample_city_renamed.to_csv(output_file, encoding="utf-8",
                    index=True, sep=",", header=True)
```
