```python
def format_date(s):
     return datetime.datetime.strptime(s, "%m/%d/%Y %H:%M")

pheonix_df.dropna(inplace=True)
pheonix_df['timestamp'] = pheonix_df['OCCURRED ON'].apply(format_date)
pheonix_df['Hour'] = pheonix_df['timestamp'].apply(lambda t: t.hour)
pheonix_df['Date'] = pheonix_df['OCCURRED ON'].apply(lambda t: t.split(' ')[0])
```

```python
crimes = pheonix_df['UCR CRIME CATEGORY'].unique()

# Add columns for each crime category, 0 = crime is not this category, 1 = crime is this category
for c in crimes:
    pheonix_df[c] = pheonix_df['UCR CRIME CATEGORY'].apply(lambda s: int(s==c))

pheonix_by_date = pheonix_df.groupby(['Date'])

pheonix_by_date[crimes].sum().head()
```

```python
pheonix_by_hour = pheonix_df.groupby(['Hour'])

pheonix_by_hour[crimes].sum().head()
```
