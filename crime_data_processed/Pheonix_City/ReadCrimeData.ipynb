```python
def format_date(s):
     return datetime.datetime.strptime(s, "%m/%d/%Y %H:%M")

pheonix_df.dropna(inplace=True)
pheonix_df['timestamp'] = pheonix_df['OCCURRED ON'].apply(format_date)
```

```python
len(pheonix_df)
```

```python
pheonix_df['timestamp'].head()
```

```python
pheonix_df['Hour'] = pheonix_df['timestamp'].apply(lambda t: t.hour)
```

```python
pheonix_by_hour = pheonix_df.groupby(['Hour', 'UCR CRIME CATEGORY'])
crime_categories = pheonix_df['UCR CRIME CATEGORY'].unique()


crime_categories_by_hour = pd.DataFrame()

for c in crime_categories:
    counts = []
    for i in range(24):
        try:
            counts.append(pheonix_by_hour['UCR CRIME CATEGORY'].count()[i][c])
        except SyntaxError:
            counts.append(0)
            
    crime_categories_by_hour[c] = counts
    
crime_categories_by_hour
```
