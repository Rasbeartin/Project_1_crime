# LA Crime Data

* Read the crime data for LA from https://data.lacity.org/A-Safe-City/Crime-Data-from-2010-to-Present/y8tr-7khq/data
* Take the data from year 2015-present
* Formatted the data with Type of crime, Date Occurred and crime count
* Store the formatted data in csv file

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python

# File to Load
raw_file = 'crime_data_LA_2015.csv'

# Read the File and store into Pandas Data Frames
la_df = pd.read_csv(raw_file)
la_df.head()
```

```python
# Read the values from Crime Code Description
cats = la_df['Crime Code Description'].unique()

cats_df = pd.DataFrame({'Crime Code Description' : cats})
#Write the Crime Code Description in csv and then work in excel to group them by primary type
cats_df.to_csv('categories.csv')
```

Edit categories_key.csv

```python
#Read the edited csv for the final category type
cat_keys = pd.read_csv('categories_key.csv')
# Mearge the original dataframe with the new categories key 
la_short_cats = pd.merge(la_df, cat_keys, on='Crime Code Description')
```

```python
la_short_cats.head()
```

```python

LA_data_short = la_short_cats.drop(['Crime Code Description', 'Status Description', 'Crime Code 1', 
                                     'Crime Code 2','Crime Code 3','Crime Code 4','Victim Descent','Premise Code',
                                    'Weapon Used Code','Weapon Description','Status Code','Status Description',
                                    'Unnamed: 2','Unnamed: 3','Cross Street','MO Codes','Area ID','Area ID','Reporting District'], axis=1)
```

```python
LA_data_short
```

```python
la_short_cats_df = pd.DataFrame(LA_data_short.groupby(["Categories Final","Date Occurred"])["Unnamed: 0"].count())
```

```python
la_short_cats_df = la_short_cats_df.rename(columns={'Unnamed: 0':'Crime_Count'}).reset_index()
```

```python
la_short_cats_df.head()
```

```python
crimes = la_short_cats_df["Categories Final"].unique()
crimes
```

```python
la_short_cats_df.to_csv("LA_Final_Crime_Data.csv",index = False)
```

```python

# Add columns for each crime category, 0 = crime is not this category, 1 = crime is this category
for c in crimes:
    la_short_cats_df[c] = la_short_cats_df['Categories Final'].apply(lambda s: int(s==c))
```

```python
la_short_cats_df.tail()
```

```python
len(la_short_cats_df['Date Occurred'].unique())
```

```python
# Group by date and export the counts to csv
LA_by_date = la_short_cats_df.groupby(['Date Occurred'])
LA_by_date[crimes].sum().to_csv('LA_by_date.csv')
```
